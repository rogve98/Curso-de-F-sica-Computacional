# Con la regla de Simpson

```python
def regla_simpson(f, a, b, n):
    """
    Calcula la integral de la función f(x) en el intervalo [a, b] utilizando la regla de Simpson.
    
    :param f: Función a integrar.
    :param a: Límite inferior del intervalo.
    :param b: Límite superior del intervalo.
    :param n: Número de subintervalos (debe ser un número par).
    :return: Valor de la integral calculado con la regla de Simpson.
    """
    if n % 2 != 0:
        raise ValueError("El número de subintervalos (n) debe ser un número par.")
    
    h = (b - a) / n
    x = [a + i * h for i in range(n + 1)]
    integral = f(a) + f(b)
    
    for i in range(1, n):
        if i % 2 == 0:
            integral += 2 * f(x[i])
        else:
            integral += 4 * f(x[i])
    
    integral *= h / 3
    return integral


def J(m, x, theta):
    """
    Integrando para las funciones de Bessel
    :Importante: La integral se hace sobre theta
    """
    c = 1 / np.pi
    j = c*np.cos(m * theta - x * np.sin(theta))
    return j


# Calcular la integral utilizando la regla de Simpson para J0(x)
def integrando_J0(theta):
    return J(0, x, theta)
    
def integrando_J1(theta):
    return J(1, x, theta)
    
def integrando_J2(theta):
    return J(2, x, theta)
```

```python
# Definir el dominio de integración
theta_min = 0  # Límite inferior de theta
theta_max = np.pi  # Límite superior de theta
num_puntos = 100 # Número de puntos de muestra en el dominio


J_0 = []
J_1 = []
J_2 = []


# Crear un arreglo de puntos de muestra en el intervalo [theta_min, theta_max]
theta_valores = np.linspace(theta_min, theta_max, num_puntos)

# Iterar a través de diferentes valores de x
for x in np.linspace(0, 10, num_puntos):
    resultado_0 = regla_simpson(integrando_J0, theta_min, theta_max, num_puntos)
    resultado_1 = regla_simpson(integrando_J1, theta_min, theta_max, num_puntos)
    resultado_2 = regla_simpson(integrando_J2, theta_min, theta_max, num_puntos)
    J_0.append(resultado_0)
    J_1.append(resultado_1)
    J_2.append(resultado_2)


# Crea una nueva figura con el tamaño especificado
plt.figure(figsize=(10, 7))

plt.plot(theta_valores, J_0, label='J0(x)')
plt.plot(theta_valores, J_1, label='J1(x)')
plt.plot(theta_valores, J_2, label='J2(x)')

plt.xlabel('x')
plt.ylabel('Valor de la integral')
plt.title('Integrales de las funciones de Bessel J0(x), J1(x), y J2(x)')
plt.legend()

# Mostrar la gráfica
plt.grid()
plt.show()
```

```python
import numpy as np
from scipy import integrate

# Definir la función J(m, x, theta)
def J(m, x, theta):
    c = 1 / np.pi
    j = c * np.cos(m * theta - x * np.sin(theta))
    return j

# Definir la función para calcular la intensidad de la luz
def intensidad(J, longitud_onda, r):
    k = (2 * np.pi) / longitud_onda
    I = ((J * (k * r)) / (k * r))**2
    return I

# Configurar los parámetros
puntos = 100  # Número de puntos
espaciado = 1  # Espaciado entre puntos

# Crear un arreglo para los resultados
grid_final = np.zeros((puntos, puntos))

# Definir el dominio de integración para theta
theta_min = 0
theta_max = np.pi
theta_valores = np.linspace(theta_min, theta_max, 1000)

# Cálculo de los valores en el arreglo
for x in range(puntos):
    for y in range(puntos):
        # Configurar la expresión matemática a calcular por punto
        r = np.sqrt(x**2 + y**2)
        
        Bessel1 = J(1,r,theta_valores)
        
        # Calcular la integral utilizando Simpson
        valor_integral = integrate.simps(Bessel1, theta_valores)
        
        # Calcular la intensidad de la luz
        intensidad_luz = intensidad((1 / np.pi) * valor_integral, 500e-9, r)
        grid_final[x, y] = intensidad_luz

# Crear un gráfico de densidad
plt.figure(figsize=(8, 8))
plt.imshow(grid_final, extent=[0, puntos, 0, puntos], cmap='hot', aspect='equal')
plt.colorbar(label='Intensidad')
plt.title('Patrón de Difracción Circular (m = 1)')
plt.xlabel('Posición en el plano focal (puntos)')
plt.ylabel('Posición en el plano focal (puntos)')
plt.show()
```

```python
# Definir la función a integrar
def f(x):
    return x**4 - 2*x + 1
```

# A lo bestia, alv

```python
# Intervalo de integración
a = 0
b = 2

# Número de "rebanadas" para N1 y N2
N1 = 10
N2 = 20

# Calcular la integral con N1 rebanadas
x1 = np.linspace(a, b, N1 + 1)
h1 = (b - a) / N1
I1 = (h1 / 2) * (f(x1[0]) + 2 * np.sum(f(x1[1:N1])) + f(x1[N1]))

# Calcular la integral con N2 rebanadas
x2 = np.linspace(a, b, N2 + 1)
h2 = (b - a) / N2
I2 = (h2 / 2) * (f(x2[0]) + 2 * np.sum(f(x2[1:N2])) + f(x2[N2]))

# Calcular el error estimado
error_estimado = (1/3) * (I2 - I1)

# Valor verdadero de la integral
valor_verdadero = 4.4

# Imprimir los resultados
print("Resultado con N1 rebanadas:", I1)
print("Resultado con N2 rebanadas:", I2)
print("Error estimado:", error_estimado)
print("Diferencia con el valor verdadero:", abs(I2 - valor_verdadero))
```

# Usando los trapezzzios

```python
from scipy.integrate import trapz

# Calcular la integral con la regla trapezoidal de SciPy
x = np.linspace(a, b, N2 + 1)
y = f(x)
resultado_scipy = trapz(y, x)

# Imprimir el resultado de SciPy y compararlo con el resultado anterior
print("Resultado de SciPy:", resultado_scipy)
print("Diferencia con el resultado anterior:", abs(resultado_scipy - I2))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special
import math

# Crear un dominio de valores x
dominio = np.linspace(-4, 4, 100)

def funcion_onda(n, x):
    c = 1 / np.sqrt((2 ** n) * np.sqrt(np.pi) * math.factorial(n))
    H = scipy.special.hermite(n)
    resultado = c * np.exp((-x ** 2) / 2) * H(x)
    return resultado

for n in [0, 1, 2, 3]:
    plt.plot(dominio, funcion_onda(n, dominio), label=f"$\Psi_{n}$")

plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import scipy.special
import scipy.integrate as spi
import math

def funcion_onda(n, x):
    c = 1 / np.sqrt((2 ** n) * np.sqrt(np.pi) * math.factorial(n))
    H = scipy.special.hermite(n)
    resultado = c * np.exp((-x ** 2) / 2) * H(x)
    return resultado

def media_cuadratica(x):
    return (x ** 2) * (funcion_onda(5, x)) ** 2

# Límites de integración finitos pero suficientemente amplios
a = -10.0
b = 10.0

# Número de puntos para la cuadratura gaussiana
N = 100

# Calcular la integral de media cuadrática utilizando cuadratura gaussiana en múltiples puntos
integral, error = spi.quadrature(media_cuadratica, a, b, maxiter=N)

print("Resultado de la integral:", np.sqrt(integral))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Lee el archivo de altitudes y almacena los datos en una matriz NumPy
data = pd.read_csv('altitudes.txt', header=None, delim_whitespace=True).values

# Define la distancia entre puntos de la cuadrícula (30,000 metros)
h = 30000

# Obtiene las dimensiones de la matriz de altitudes
filas, columnas = data.shape

# Inicializa matrices para almacenar las derivadas ∂w/∂x y ∂w/∂y
derivada_x = np.zeros((filas, columnas))
derivada_y = np.zeros((filas, columnas))

# Calcula las derivadas ∂w/∂x y ∂w/∂y en cada punto de la cuadrícula
for i in range(filas):
    for j in range(columnas):
        # Aproximación de la derivada ∂w/∂x (diferencia hacia adelante)
        if j < columnas - 1:
            derivada_x[i, j] = (data[i, j + 1] - data[i, j]) / h
        
        # Aproximación de la derivada ∂w/∂y (diferencia hacia adelante)
        if i < filas - 1:
            derivada_y[i, j] = (data[i + 1, j] - data[i, j]) / h

# Ángulo φ en radianes (45 grados)
phi = np.deg2rad(45)

# Calcula la intensidad para cada punto de la cuadrícula
intensidad = (np.cos(phi) * derivada_x + np.sin(phi) * derivada_y) / np.sqrt(derivada_x**2 + derivada_y**2 + 1)

# Crea un gráfico de densidad (mapa en relieve) de los valores de intensidad
plt.figure(figsize=(10, 8))
plt.imshow(intensidad, cmap='gray', extent=(0, columnas, 0, filas), origin='upper')
plt.colorbar(label='Intensidad')
plt.title('Mapa en Relieve de Altitudes')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.show()
```
