```julia
using Plots

# Tamao de la cuadrcula
N = 800

# Rango de valores para x e y
x_min, x_max = -2.0, 2.0
y_min, y_max = -2.0, 2.0

# Crear una cuadrcula de valores complejos en el rango dado
x = range(x_min, x_max, length=N)
y = range(y_min, y_max, length=N)
c_values = [x[i] + y[j] * im for j in 1:N, i in 1:N]

# Inicializar la matriz de salida con ceros (todos los puntos fuera del conjunto)
mandelbrot = zeros(UInt8, N, N)

# Nmero mximo de iteraciones
max_iterations = 100

# Iterar para cada punto c
z = zeros(Complex{Float64}, N, N)
for i in 1:max_iterations
    z .= z.^2 .+ c_values
    # Verificar si los puntos han escapado
    mask = abs.(z) .< 2.0
    mandelbrot .+= mask
end

# Crear la imagen de densidad (density plot)
heatmap(x, y, mandelbrot, color=:grays, axis=false)
title!("Conjunto de Mandelbrot")
xlabel!("Re(x)")
ylabel!("Im(y)")
```

```julia
using Plots

# Tamao de la cuadrcula
N = 800

# Rango de valores para x e y
x_min, x_max = -2.0, 2.0
y_min, y_max = -2.0, 2.0

# Crear una cuadrcula de valores complejos en el rango dado
x = range(x_min, x_max, length=N)
y = range(y_min, y_max, length=N)
c_values = [x[i] + y[j] * im for j in 1:N, i in 1:N]

# Inicializar la matriz de salida con ceros (todos los puntos fuera del conjunto)
mandelbrot = zeros(UInt8, N, N)

# Nmero mximo de iteraciones
max_iterations = 100

# Iterar para cada punto c y registrar el nmero de iteraciones
z = zeros(Complex{Float64}, N, N)
for i in 1:max_iterations
    z .= z.^2 .+ c_values
    # Verificar si los puntos han escapado
    mask = abs.(z) .< 2.0
    mandelbrot .+= mask
end

# Colorear los puntos segn el nmero de iteraciones
heatmap(x, y, mandelbrot, color=:jet,  axis=false, legend=:none)
title!("Conjunto de Mandelbrot (Coloreado por Nmero de Iteraciones)")
xlabel!("Re(x)")
ylabel!("Im(y)")
```
