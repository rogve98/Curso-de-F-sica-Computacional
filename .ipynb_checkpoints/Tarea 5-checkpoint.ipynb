## Integración trapezoidal

$$
\int_a^b f(x)\, dx=\frac{b-a}{n}\left(\frac{f(a)+f(b)}{2}+\sum_{k=1}^{n-1}f\left(a+k\cdot\frac{b-a}{n}\right)\right)
$$

```python
ruta = "velocidades.txt"
vel = open(ruta,'r')
lineas =  vel.readlines()
vel.close()
```

```python
lineas
velocidades = []
tiempos = []
for linea in lineas:
    t,v = linea.split('\t')
    velocidades.append(float(v))
    tiempos.append(float(t))
```

```python
plt.plot(tiempos,velocidades)
plt.title("Velocidades")
plt.ylabel("Distancia")
plt.xlabel("Tiempo")
```

```python
def vel(i):
    return velocidades[int(i)]
```

```python
#Integración trapecio
def trapecio(f,a,b,n):
    h = (b-a)/n
    x = [a+h*i for i in range(0,n+1)]
    sum = 0
    for i in range(1,n):
        sum += f(x[i])
        #sum += f(a+i*(b-a)/n)

    return (b-a)/n*((f(a)+f(b))/2+sum)
```

```python
trapecio(vel,0,100,100)
```

```python
def posiciones(tiempos):
    pos = [0]
    sum = 0
    for i in range(0,100):
        sum += trapecio(vel,i,i+1,1)
        pos.append(sum)
    return pos
```

```python
fig,ax=plt.subplots(figsize=(10,6))

plt.title("Velocidad y tiempos recorridos")

#Con este método encimamos gráficas con diferentes ylabel
ax2 = ax.twinx()
ax.plot(tiempos, posiciones(tiempos), color = 'r')
ax2.plot(tiempos, velocidades, color = 'b')
 
# giving labels to the axises
ax.set_xlabel('Tiempo (s)')#eje x
ax.set_ylabel('Distancia recorrida (m)', color = 'r')#eje y 1: distancia en m
ax2.set_ylabel('Velocidad (m/s)', color = 'b')#eje y 2: velocidad en m/s
plt.tight_layout()
ax.grid()
plt.show()
```

## Integración método de Simpson

\begin{align}
\int_a^b f(x)\, dx&=\frac{h}{3}[f(x_0)+4f(x_1)+2f(x_2)+4f(x_3)+2f(x_2)+…+2f(x_{n-2})+4f(x_{n-1})+f(x_n)]\\
&=\sum_{j\ \text{par}}^{n/2} 2f(x_{2i})+\sum_{i\ \text{impar}}^{n/2+1} 4f(x_{2i-1})+ f(x_0)-f(x_n)
\end{align}

```python
#Ejercicio 2
def func(x): return x**4-2*x+1
```

```python
#Simpson
def simpson(f,a,b,n):
    h = (b-a)/n
    sum = 0
    S0 = f(a)-f(b)
    #x = [a+i*h for i in range(0,101)]
    for i in range(1,int(n/2)+1):
        sum += 4*f(a+(2*i-1)*h)+2*f(a+(2*i)*h)
    return h/3*(sum+S0)
```

```python
#Inciso a)
trap = trapecio(func,0,2,10)
simp = simpson(func,0,2,10)
print(Fore.GREEN+Style.BRIGHT,"Trapecio:",trap,"  Simpson:",simp)
```

```python
#b)Error numérico
print("Trapecio :",abs(trap-4.4))
print("Simpson: ",abs(simp-4.4))
```

## Límite de difracción de un telescopio

La intensidad de la luz en el patrón de difracción dado por un telescopio circular esta dada por:
$$
I(r)=\left(\frac{J_1(kr)}{kr)}\right)^2
$$
donde $r$ es la distancia en el plano focal desde el centro del patrón de difracción $k=2\pi/\lambda$ y $J_1(x)$ es una función de Bessel. Las funciones de Bessel $J_m(x)$ se definen de la siguiente manera:
$$
J_m(x)=\frac{1}{\pi}\int_0^\pi\cos(m\theta-x\sin\theta)\ d\theta
$$
donde $m$ es un entero no negativo y $x\geq 0$

```python
#Inciso a)
def J(m,x):
    def f(θ):
        return np.cos(m*θ-x*np.sin(θ))
    return (1/np.pi)*simpson(f,0,np.pi,1000)
```

```python
x = np.linspace(0,20,100)
sol0 = []
sol1 = []
sol2 = []
for i in x:
    r0 = J(0,i)
    r1 = J(1,i)
    r2 = J(2,i)
    sol0.append(r0)
    sol1.append(r1)
    sol2.append(r2)
    

plt.plot(x,sol0)
plt.plot(x,sol1)
plt.plot(x,sol2)
```

```python
#Inciso b)
def j(m,x):
    def f(θ):
        return np.cos(m*θ-x*np.sin(θ))
    #Alternativa interesante con λ: 
    #f = lambda θ: np.cos(m*θ-x*np.sin(θ))
    return f
    
θ = np.linspace(0,np.pi,1000)
j0 = []
j1 = []
j2 = []
for i in x:
    j0.append(1/np.pi*integrate.simpson(j(0,i)(θ),x=θ))#Se solucionó la advertencia pasando a θ como argumento con palabra clave
    j1.append(1/np.pi*integrate.simpson(j(1,i)(θ),x=θ))#Es decir, x=θ, siendo que θ es el dominio de integración de la función
    j2.append(1/np.pi*integrate.simpson(j(2,i)(θ),x=θ))#de la función cos(...)
    
plt.plot(x,j0)
plt.plot(x,j1)
plt.plot(x,j2)
```

```python
#Ejercicio 5
def Fgrav(x,y,z):
     return z / (x**2 + y**2 + z**2)**(3/2)

def cuadratura_gauss_2d(f, limites_u, limites_v, n,z):
    # Obtener puntos y pesos de Gauss-Legendre
    #gauss_p, pesos = special.roots_legendre(n)
    gauss_p, pesos = np.polynomial.legendre.leggauss(n)
    
    # Cambiar los puntos y pesos a los límites deseados
    u_mid, u_range = (limites_u[1] + limites_u[0]) / 2, (limites_u[1] - limites_u[0]) / 2
    v_mid, v_range = (limites_v[1] + limites_v[0]) / 2, (limites_v[1] - limites_v[0]) / 2
    puntos_u = u_mid + u_range * gauss_p
    puntos_v = v_mid + v_range * gauss_p
    pesos_u = u_range * pesos
    pesos_v = v_range * pesos

    # Inicializar la integral
    integral = 0.0

    # Realizar la doble suma para la cuadratura de Gauss
    for i in range(n):
        for j in range(n):
            integral += pesos_u[i] * pesos_v[j] * f(puntos_u[i], puntos_v[j],z)
    
    return integral

G = 6.67430e-11 
L = 10 #metros
σ = 10_000/(L**2)# = 100 kg/m^2
n = 150 #puntos de gauss

limites_u = (-L/2,L/2)
limites_v = (-L/2,L/2)

zs = np.linspace(0.1,10,100)
Fz = [G*σ*cuadratura_gauss_2d(Fgrav,limites_u,limites_v,n,z) for z in zs]

plt.figure(figsize=(10,7))

plt.plot(zs, Fz, 'orange')
plt.xlabel("z (m)")
plt.ylabel("Fz (N)")
plt.title("Fuerza en función de z")
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.integrate import dblquad
import matplotlib.pyplot as plt

# Constantes
G = 6.67430e-11  # Constante de gravitación universal en m^3 kg^-1 s^-2
sigma = 100.0      # Densidad superficial de masa (en kg/m^2)
L = 10.0          # Lado de la placa (en metros)

# Función integrando (parte de la fórmula de la fuerza)
def integrand(y, x, z):
    return 1.0 / (x**2 + y**2 + z**2)**(3/2)

# Límites de integración para x y y (la placa es cuadrada)
x_min = -L/2
x_max = L/2
y_min = -L/2
y_max = L/2

# Función para calcular la fuerza F_z para un valor dado de z
def F_z_at_z(z):
    # Realizamos la integración doble
    integral, error = dblquad(integrand, x_min, x_max, lambda x: y_min, lambda x: y_max, args=(z,))
    # Fuerza en la dirección z
    return G * sigma * z * integral

# Definir el rango de z (de 0 a 10 metros)
z_values = np.linspace(0.1, 10, 100)  # Evitamos z=0 para evitar singularidad
F_z_values = [F_z_at_z(z) for z in z_values]  # Calcular F_z para cada z

# Graficar F_z en función de z
plt.figure(figsize=(8, 6))
plt.plot(z_values, F_z_values, label="Fuerza Gravitacional $F_z$")
plt.xlabel('Distancia $z$ (m)')
plt.ylabel('Fuerza $F_z$ (N)')
plt.title('Fuerza Gravitacional $F_z$ en función de $z$')
plt.grid(True)
plt.legend()
plt.show()
```
