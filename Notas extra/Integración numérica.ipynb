# Ejercicio 1: Integración por trapecios.

```python
# Ejercicio 1
ruta = "velocidades.txt"
vel = open(ruta,'r')
lineas =  vel.readlines()
vel.close()
```

```python
lineas
velocidades = []
tiempos = []
for linea in lineas:
    t,v = linea.split('\t')
    velocidades.append(float(v))
    tiempos.append(float(t))
```

Mi forma de resolverlo fue la siguiente, defino una función que me devuelva cada entrada del vector de velocidades. La idea de hacerlo función es para poderlo meter en `trapecio()` que es una forma general de aplicar el método, únicamente hay que pasarle la función a integrar, los límites y el número de cachos en los que vamos a partir el intervalo.

Para ello seguimos la regla del trapecio:
$$
\int_a^b f(x)\, dx=\frac{b-a}{n}\left(\frac{f(a)+f(b)}{2}+\sum_{k=1}^{n-1}f\left(a+k\cdot\frac{b-a}{n}\right)\right)
$$
Noten que he generado una lista `x` en donde se define cada argumento de $f$ considerando la suma desde $k$ hasta $n-1$. Esto es sumamente conveniente  pues nada más hay que pasarle cada elemento de esa lista a la función, evaluarla y aplicar la suma. Dejé comentada otra forma de aplicar `sum` sin considerar la lista `x`.

```python
def vel(i):
    return velocidades[int(i)]
#Integración trapecio
def trapecio(f,a,b,n):
    h = (b-a)/n
    x = [a+h*k for k in range(0,n+1)]
    sum = 0
    for i in range(1,n):
        sum += f(x[i])
        #sum += f(a+i*(b-a)/n)

    return (b-a)/n*((f(a)+f(b))/2+sum)

def posiciones(tiempos):
    pos = [0]
    sum = 0
    for i in range(0,100):
        sum += trapecio(vel,i,i+1,1)
        pos.append(sum)
    return pos
```

```python
#a)
trapecio(vel,0,100,100)
```

```python
#b)
fig,ax=plt.subplots(figsize=(10,6))

plt.title("Velocidad y tiempos recorridos")

#Con este método encimamos gráficas con diferentes ylabel
ax2 = ax.twinx()
ax.plot(tiempos, posiciones(tiempos), color = 'r')
ax2.plot(tiempos, velocidades, color = 'b')
 
ax.set_xlabel('Tiempo (s)')#eje x
ax.set_ylabel('Distancia recorrida (m)', color = 'r')#eje y 1: distancia en m
ax2.set_ylabel('Velocidad (m/s)', color = 'b')#eje y 2: velocidad en m/s
plt.tight_layout()
ax.grid()
plt.show()
```

# Ejercicio 2: Integración por Simpson

**Integración por regla de Simpson**

\begin{align}
\int_a^b f(x)\, dx&=\frac{h}{3}[f(x_0)+4f(x_1)+2f(x_2)+4f(x_3)+2f(x_2)+…+2f(x_{n-2})+4f(x_{n-1})+f(x_n)]\\
&=\sum_{j\ \text{par}}^{n/2} 2f(x_{2i})+\sum_{i\ \text{impar}}^{n/2+1} 4f(x_{2i-1})+ f(x_0)-f(x_n)
\end{align}

```python
#Ejercicio 2
def func(x): return x**4-2*x+1
#Simpson
def simpson(f,a,b,n):
    h = (b-a)/n
    sum = 0
    S0 = f(a)-f(b)
    #x = [a+i*h for i in range(0,101)]
    for i in range(1,int(n/2)+1):
        sum += 4*f(a+(2*i-1)*h)+2*f(a+(2*i)*h)
    return h/3*(sum+S0)
```

Comenzamos nuevamente con el paso de integración h y definiendo la función evaluada en los extremos del intervalo considerado. En este caso no ocuparemos la lista (aunque lo dejé comentado por cualquier cosa). Vamos a considerar solamente un ciclo for y vamos a aplicar la fórmula oportunamente considerando índices pares e impartes, por lo tanto el `for` puede correr de $1$ a $n/2+1$. La suma par se múltiplica por un índice par `2*i` y la suma impar por un índice `2*i-1` y todo ello considerando a `h` su paso de integración.

```python
#Inciso a)
trap = trapecio(func,0,2,10)
simp = simpson(func,0,2,10)
print("Trapecio:",trap,"  Simpson:",simp)
```

# Ejercicio 3: Límite de difracción de un telescopio

La intensidad de la luz en el patrón de difracción dado por un telescopio circular esta dada por:
$$
I(r)=\left(\frac{J_1(kr)}{kr)}\right)^2
$$
donde $r$ es la distancia en el plano focal desde el centro del patrón de difracción $k=2\pi/\lambda$ y $J_1(x)$ es una función de Bessel. Las funciones de Bessel $J_m(x)$ se definen de la siguiente manera:
$$
J_m(x)=\frac{1}{\pi}\int_0^\pi\cos(m\theta-x\sin\theta)\ d\theta
$$
donde $m$ es un entero no negativo y $x\geq 0$

```python
#Inciso a)
def J(m,x):
    def f(θ):
        return np.cos(m*θ-x*np.sin(θ))
    return (1/np.pi)*simpson(f,0,np.pi,1000)
```

# Ejercicio 4: Integración de gauss 1D

$$
\int_{-1}^1 f(x)\, dx=\sum_{i=1}^n w_i\cdot f(x_i)
$$

donde los $w_i$ y $x_i$ son las raíces (pesos y puntos) de los polinomios de legendre. Para ello no olvidar el siguiente cambio de variable:

$$
\int_a^b f(x)\, dx=\frac{b-a}{2}\int_{-1}^1 f\left (\frac{b-a}{2}u+\frac{a+b}{2}\right)
$$

Aplicamos cuadtatura de gauss del lado derecho de esta integral (el cambio de variable) de modo que

$$
\int_a^b f(x)\, dx\approx \frac{b-a}{2}\sum_{i=1}^n w_i\cdot f\left(\frac{b-a}{2}u_i+\frac{a+b}{2} \right)
$$

```python
N=50
a=0
V=1/1000
p=6.022e28
k= 1.380649e-23
def gauss_xw_ab(N,a,b):
    x,w = special.roots_legendre(N)
    return 0.5*(b-a)*x+0.5*(b+a),0.5*(b-a)*w
    
def f(x): return (x**4 * np.exp(x))/((np.exp(x)-1)**2)

def Cv(b):
    x,w = gauss_xw_ab(N,a,428/b)
    S=0
    for i in range(N):
        S = S + w[i]*f(x[i])
        
    C=9*V*p*k*(b/428)**3 * S
    return C

print(Cv(100))
```

# Ejercicio 5: Integración de gauss 2D

Aquí solo hay que extender la cuadratura a dos variables de una función $f:\mathbb{R}^2\to\mathbb{R}$.

```python
#Ejercicio 5
def Fgrav(x,y,z):
     return z / (x**2 + y**2 + z**2)**(3/2)

"""
f := función a integrar de R2 en R
limites_u := una tupla de dos entradas para los límites de integración en x
limites_c := una tupla de dos entradas para los límites de integración en y
n := puntos de gauss
z := variable z de la fuerza Fgrav()
"""
def cuadratura_gauss_2d(f, limites_u, limites_v, n,z):
    #gauss_p, pesos = special.roots_legendre(n)
    gauss_p, pesos = np.polynomial.legendre.leggauss(n)
    
    # Cambiar los puntos y pesos a los límites deseados
    u_mid, u_range = (limites_u[1] + limites_u[0]) / 2, (limites_u[1] - limites_u[0]) / 2
    v_mid, v_range = (limites_v[1] + limites_v[0]) / 2, (limites_v[1] - limites_v[0]) / 2
    puntos_u = u_mid + u_range * gauss_p
    puntos_v = v_mid + v_range * gauss_p
    pesos_u = u_range * pesos
    pesos_v = v_range * pesos

    # Inicializar la integral
    integral = 0.0

    # Realizar la doble suma para la cuadratura de Gauss
    for i in range(n):
        for j in range(n):
            integral += pesos_u[i] * pesos_v[j] * f(puntos_u[i], puntos_v[j],z)
    
    return integral

G = 6.67430e-11 
L = 10 #metros
σ = 10_000/(L**2)# = 100 kg/m^2
n = 150 #puntos de gauss

limites_u = (-L/2,L/2)
limites_v = (-L/2,L/2)

#Definimos una epsilon para salvarnos de la singularidad para x=y=z=0
ϵ = 0.1
zs = np.linspace(0,10,100)
Fz = [G*σ*cuadratura_gauss_2d(Fgrav,limites_u,limites_v,n,z+ϵ) for z in zs]

plt.figure(figsize=(10,7))

plt.plot(zs, Fz, 'orange')
plt.xlabel("z (m)")
plt.ylabel("Fz (N)")
plt.title("Fuerza en función de z")
plt.grid(True)
plt.show()
```
